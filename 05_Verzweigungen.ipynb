## Verzweigungen in Python

### Bedingungen

Bei einer Verzweigung wird eine Bedingung überprüft, die entweder erfüllt ist oder nicht. Übliche Operatoren dafür sind: 

| Operator    | Name                   |
|-------------|------------------------|
|     ==      |     Gleich             |
|     !=      |     Ungleich           |
|     <       |     Kleiner            |
|     >       |     Größer             |
|     <=      |     Kleiner gleich     |
|     >=      |     Größer gleich      |
|     and      |     <font color="white" >UND-Verknüpfung bei mehreren Bedingungen </font>                |
|     or    |     ODER-Verknüpfung bei mehreren Bedingungen               |

### Mögliche Ergebnisse einer Bedingung: 
wahr = True  
falsch = False


### Aufgabe: Sind die Bedingungen erfüllt oder nicht?
`zahl1=5`  
`zahl2=10`  
`zahl3=10`  
`zahl4=15`  
`zahl5=9`  

|Nummer| Bedingung             	| Ergebnis 	|
|------|-----------------------	|----------	|
|1|     `zahl2== zahl3`   	|     `True`     	|
|2|     `zahl2== zahl4`                	|          	|
|3|     `zahl2< zahl4`              	|          	|
|4|      `zahl2> zahl4`            	|          	|
|5|     `zahl2>= zahl5`                 	|          	|
|6|     `zahl2<= zahl3`                	|          	|
|7|    `zahl2<= zahl3 and zahl4<= zahl3`                 	|          	|
|8|    `zahl2<= zahl3 or zahl4<= zahl3`               	|          	|


```python
## Lösung Aufgabe 1

zahl1=5
zahl2=10
zahl3=10
zahl4=15
zahl5=9

## 1
if  zahl2== zahl3:
    print ("True")  
else:
    print ("False")
## 2
if  zahl2== zahl4:
    print ("True")  
else:
    print ("False")
## 3
if  zahl2< zahl4:
    print ("True")  
else:
    print ("False")
## 4
if  zahl2> zahl4:
    print ("True")  
else:
    print ("False")

## 5
if  zahl2>= zahl5:
    print ("True")  
else:
    print ("False")

## 6
if  zahl2<= zahl3:
    print ("True")  
else:
    print ("False")

## 7
if  zahl2<= zahl3 and zahl4<= zahl3:
    print ("True")  
else:
    print ("False")

## 8
if  zahl2<= zahl3 or zahl4<= zahl3:
    print ("True")  
else:
    print ("False")
 
```

## Syntax von Verzweigungen: 

`if  Bedingung:  `<br>
  <font color="white"> ......</font> `Dann-Anweisung `  <br>
`else: ` <br>
<font color="white">......</font>`Sonst-Anweisung`<br>  


**Wichtig dabei sind alle Zeichen, auch die Einrückungen!**  


Wenn die Bedingung erfüllt ist, dann wird die Dann-Anweisung ausgeführt. Ansonsten wird die Sonst-Anweisung nach dem `else: ` ausgeführt.   


Der else-Block kann auch weggelassen werden.

Falls mehrere Bedingungen abgefragt werden sollen, kann die Syntax auch so aussehen: 

`if  Bedingung:  `<br>
  <font color="white"> ......</font> `Dann-Anweisung `  <br>
`elif Bedingung: ` <br>
<font color="white">......</font>`Dann-Anweisung `  <br>
`else: ` <br>
<font color="white">......</font>`Sonst-Anweisung`<br>
